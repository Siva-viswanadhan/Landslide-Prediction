first, landslide type prediction

```python
df1=df
```

```python
df1.dtypes
```

encoding

```python
dfe1=pd.get_dummies(df[['infrastructure_quality','season']],drop_first=True,dtype=int)
dfe1
```

concatenate

```python
df_1=pd.concat([df1,dfe1],axis=1)
df_1
```

dropping

```python
df_1=df_1.drop(['population_density','hour','day_night','fatality_level','infrastructure_quality','season','previous_landslides','rescue_response_level'],axis=1)
df_1
```

data x,y input ,output


```python
x1=df_1.drop('landslide_type',axis=1)
x1
```

```python
y1=df_1['landslide_type']
y1
```

test,train split

```python
from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.30,random_state=1)
```

model training

```python
from sklearn.ensemble import RandomForestClassifier
model1=RandomForestClassifier(n_estimators=100,
    criterion='entropy',
    random_state=42,
    n_jobs=-1)
model1.fit(x1_train,y1_train)
```

```python
y1_pred=model1.predict(x1_test)
y1_pred
```

```python
y1_test
```

```python
from sklearn.metrics import accuracy_score
score1=accuracy_score(y1_test,y1_pred)
score1
```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)
cv_scores = cross_val_score(model1, x1, y1, cv=cv, scoring='accuracy')
cv_scores
```

```python
import pickle
from pathlib import Path

out_dir = Path("models")
out_dir.mkdir(exist_ok=True)

models = {
    "landslide_type_model.pkl": model1,
    "fatality_model.pkl": model2,
    "rescue_response_model.pkl": model3,
}

for fname, mdl in models.items():
    with open(out_dir / fname, "wb") as f:
        pickle.dump(mdl, f, protocol=pickle.HIGHEST_PROTOCOL)

print(f"Saved {len(models)} models to: {out_dir.resolve()}")
```
